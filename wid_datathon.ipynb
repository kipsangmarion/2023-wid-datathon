## 7. Literature Review

### Climate Finance and Global Disparities in Climate Finance Allocation

Climate finance, a crucial component of the global response to climate change, involves the mobilization and distribution of financial resources to address both mitigation and adaptation efforts. The allocation of climate finance has garnered significant attention, with scholars emphasizing the need for equitable distribution to ensure effective climate action (Buchner et al., 2017).

While substantial efforts have been made to understand the overall allocation of climate finance, there exists a gap in the literature concerning the specific distribution patterns related to early warning systems (EWS) and climate services. The global population's vulnerability to extreme weather events underscores the importance of understanding how financial resources are directed towards projects that enhance resilience and preparedness.

### Disparities in Early Warning Systems Coverage

Studies highlight the inadequacy of global coverage provided by Early Warning Systems (EWS), with nearly two-thirds of the world's population lacking access to timely warnings (Basher, 2006). This significant gap in coverage increases vulnerability, particularly in developing regions, and emphasizes the need for targeted investments in EWS (UNDRR, 2019).

The literature suggests that timely warnings can reduce damages caused by natural disasters by up to 30%, making a compelling case for increased investments in EWS as a cost-effective strategy (Stern Review, 2006). Despite the recognized benefits, there is a notable lack of literature exploring the disparities in climate finance allocation specifically for EWS projects.

### Paris Agreement and Science-Based Climate Action

The Paris Agreement, a landmark accord in the realm of climate change, emphasizes the importance of science-based climate action. The agreement highlights the necessity of adapting to the adverse impacts of climate change, with a specific focus on enhancing resilience and reducing vulnerability (United Nations, 2015). This global commitment necessitates a comprehensive analysis of climate finance allocation, with a particular emphasis on projects supporting EWS and climate services aligned with science-based strategies.



### Trends in Project Numbers and Regional Disparities

The observed upward trend in the number of projects related to EWS and climate services aligns with broader discussions on the increasing importance of climate adaptation and resilience-building initiatives (Klein et al., 2014). The literature further suggests that understanding regional disparities in climate finance distribution is vital for designing targeted interventions (Berrang-Ford et al., 2011).

### Disparity Between Climate Services and Early Warning Systems

The literature on climate finance often explores the allocation of funds to broader categories such as adaptation and mitigation. However, there is a scarcity of specific studies addressing the observed disparity between funding for climate services and early warning systems. This literature gap highlights the novelty and importance of the current analysis.

## 10. Key Findings

### Finding 1: Number of projects on climate services and early warning systems over the years

The following are notable observations regarding the number of projects:
- The number of projects has experienced an upward trend between 1989 and 2023.
- This increase is concurrent with a rise in financial investments directed towards climate services and early warning systems.
- The health sector has experienced the most significant growth in total project expenditure, with the most significant spike occurring in 2020, largely attributed to the impact of the COVID-19 pandemic.


```python
# Finding #1: Number of projects over the years
data['Board Approval Date'] = pd.to_datetime(data['Board Approval Date'])
data['Year'] = data['Board Approval Date'].dt.year
projects_over_time = data.groupby('Year')['Project Id'].count()

# Plotting the number of projects over the years
plt.figure(figsize=(10, 6))
sns.lineplot(x=projects_over_time.index, y=projects_over_time.values)
plt.title('Number of Projects Over the Years')
plt.xlabel('Year')
plt.ylabel('Number of Projects')
plt.show()
```

```python
# Group by 'Year' and 'Sector 1', then sum the 'Total Project Cost $US'
sector_costs = data.groupby(['Year', 'Sector 1'])['Total Project Cost $US'].sum().reset_index()

# Select the top 5 sectors based on the total project cost
top_sectors = sector_costs.groupby('Sector 1')['Total Project Cost $US'].sum().nlargest(5).index

# Filter the dataframe for the top 5 sectors
top_sectors_data = sector_costs[sector_costs['Sector 1'].isin(top_sectors)]

# Create a line plot for each sector with different colors
plt.figure(figsize=(14, 8))
sns.lineplot(x='Year', y='Total Project Cost $US', hue='Sector 1', data=top_sectors_data)
plt.title('Sum of Total Project Costs Over the Years for Top 10 Sectors')
plt.xlabel('Year')
plt.ylabel('Total Project Cost $US')
plt.legend(title='Sector')
plt.show()
```

### Finding 2: Regional distribution of projects on early warning systems and climate services

- The total project costs across all nine regions ranged from 13,479,463 USD to 168,770,034,685 USD, with the East Asia and Pacific region receiving the highest funding, accounting for 22.5% of the total funds. On the other hand, most African regions received the least amount of funding.
- Of the 10 countries that received the highest funding, only two were from the African region, which received almost half of what the most funded country received.

```python
# Group by 'Region' and sum the 'Total Project Cost $US'
region_total_costs = data.groupby('Region')['Total Project Cost $US'].sum().reset_index()

# Create a bar plot for each region
plt.figure(figsize=(12, 6))
sns.barplot(x='Region', y='Total Project Cost $US', data=region_total_costs)
plt.title('Total Project Costs by Region')
plt.xlabel('Region')
plt.ylabel('Total Project Cost $US')
plt.xticks(rotation=45, ha='right')
plt.show()
```

```python
# Group by 'Country' and sum the 'Total Project Cost $US'
country_total_costs = data.groupby('Country')['Total Project Cost $US'].sum().reset_index()

# Select the top 10 countries based on the total project cost
top_countries = country_total_costs.groupby('Country')['Total Project Cost $US'].sum().nlargest(10).index

# Filter the dataframe for the top 10 countries
top_countries_data = country_total_costs[country_total_costs['Country'].isin(top_countries)]

# Create a bar plot for each country
plt.figure(figsize=(12, 6))
sns.barplot(x='Country', y='Total Project Cost $US', data=top_countries_data)
plt.title('Total Project Costs by Country (Top 10)')
plt.xlabel('Country')
plt.ylabel('Total Project Cost $US')
plt.xticks(rotation=45, ha='right')
plt.show()
```

### Finding 3: Disparity between projects on early warning systems and those on climate services

- Climate Services accounted for 95.67% of the total project cost of $US.
- The sum of the total project cost for Climate Services (728829115579) exceeded that of Early Warning Systems (33013199106).

```python
# Group by 'Category' and sum the 'Total Project Cost $US'
category_total_costs = data.groupby('Category')['Total Project Cost $US'].sum().reset_index()

# Create a bar plot for each category
plt.figure(figsize=(12, 6))
sns.barplot(x='Category', y='Total Project Cost $US', data=category_total_costs)
plt.title('Total Project Costs by Category')
plt.xlabel('Category')
plt.ylabel('Total Project Cost $US')
plt.xticks(rotation=45, ha='right')
plt.show()
```

- Compared to other regions, South Asia receives more funding for their early warning systems.
- Despite having the highest climate service funding, the East Asia and Pacific Region receive less funding towards early warning systems.
- Both categories receive less funding in African regions.

```python
# Filter the dataframe for projects in the 'Climate Services' category
climate_services_data = data[data['Category'] == 'Climate Services']

# Group by 'Region' and sum the 'Total Project Cost $US'
region_total_costs = climate_services_data.groupby('Region')['Total Project Cost $US'].sum().reset_index()

# Create a bar plot for each region
plt.figure(figsize=(12, 6))
sns.barplot(x='Region', y='Total Project Cost $US', data=region_total_costs)
plt.title('Total Project Costs for Climate Services by Region')
plt.xlabel('Region')
plt.ylabel('Total Project Cost $US')
plt.xticks(rotation=45, ha='right')
plt.show()
```

```python
# Filter the dataframe for projects in the 'Early Warning Systems' category
ews_data = data[data['Category'] == 'Early Warning Systems']

# Group by 'Region' and sum the 'Total Project Cost $US'
region_ews_total_costs = ews_data.groupby('Region')['Total Project Cost $US'].sum().reset_index()

# Create a bar plot for each region
plt.figure(figsize=(12, 6))
sns.barplot(x='Region', y='Total Project Cost $US', data=region_ews_total_costs)
plt.title('Total Project Costs for Early Warning Systems by Region')
plt.xlabel('Region')
plt.ylabel('Total Project Cost $US')
plt.xticks(rotation=45, ha='right')
plt.show()
```

### Finding 4:   Status of the Projects by Region and Financing Type

- It's worth noting that Europe and Central Asia have executed the most projects, while Eastern and Southern Africa have the greatest number of ongoing projects.
- Also, most of the projects financed by Grants are active, while most projects financed by IBRD are closed.

```python
# Group by 'Region' and 'Project Status' and get the counts
region_project_status_counts = data.groupby(['Region', 'Project Status']).size().unstack(fill_value=0)

# Create a bar plot
plt.figure(figsize=(12, 6))
region_project_status_counts.plot(kind='bar', stacked=False)
plt.title('Project Status Counts by Region')
plt.xlabel('Region')
plt.ylabel('Count')
plt.xticks(rotation=45, ha='right')
plt.legend(title='Project Status', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```

```python
# Calculate the percentage of project status by financing type
status_financing_percentage = data.groupby(['Project Status', 'Financing Type']).size().unstack(fill_value=0).apply(lambda x: x / x.sum(), axis=1) * 100

# Create a bar plot
plt.figure(figsize=(12, 6))
status_financing_percentage.plot(kind='bar', stacked=False, cmap='viridis')
plt.title('Percentage of Project Status by Financing Type')
plt.xlabel('Project Status')
plt.ylabel('Percentage')
plt.xticks(rotation=45, ha='right')
plt.legend(title='Financing Type', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```
